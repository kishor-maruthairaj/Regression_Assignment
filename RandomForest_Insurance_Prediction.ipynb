```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset.head(5)
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset.head(5)
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.20,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=2500,criterion='absolute_error')
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
import pickle
filename="final_RandomForest_Insurance_Prediction.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("final_RandomForest_Insurance_Prediction.sav",'rb'))
result=loaded_model.predict([[1234,345,4565,1,0]])
```
